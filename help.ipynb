```python
import numpy as np

def estimate_price_by_roe(bps0, roe, ke, years=5, g=0.02):
    """
    使用简化剩余收益模型估算股价
    参数：
    bps0: 当前每股净资产
    roe: 预测的未来ROE（假设前几年恒定）
    ke: 股东必要报酬率
    years: 高速预测年数
    g: 永续增长率
    """
    estimated_ri_pv = 0  # 剩余收益现值
    bps = bps0

    # 计算预测期内的剩余收益现值
    for t in range(1, years + 1):
        ri = bps * (roe - ke)  # 当年剩余收益
        ri_pv = ri / ((1 + ke) ** t)  # 折现
        estimated_ri_pv += ri_pv
        bps = bps * (1 + roe)  # 假设利润全留存，净资产增长

    # 计算永续终端价值（预测期后）
    ri_terminal = bps * (roe - ke) * (1 + g)
    terminal_value_pv = ri_terminal / ((ke - g) * ((1 + ke) ** years))

    # 内在价值 = 当前净资产 + 剩余收益现值总和
    intrinsic_value = bps0 + estimated_ri_pv + terminal_value_pv
    return intrinsic_value

# 示例：假设一家公司
bps0 = 6.91 # 当前每股净资产10元
roe = 0.17  # 预测未来ROE为15%
ke = 0.10    # 股东要求10%的回报率
g = 0.04
estimated_price = estimate_price_by_roe(bps0, roe, ke, g=g)
print(f"基于模型估算的每股内在价值约为：{estimated_price:.2f} 元")
```

```python
from datetime import datetime
import backtrader as bt

class SmaCross(bt.SignalStrategy):
    def __init__(self):
        sma1, sma2 = bt.ind.SMA(period=10), bt.ind.SMA(period=30)
        crossover = bt.ind.CrossOver(sma1, sma2)
        self.signal_add(bt.SIGNAL_LONG, crossover)

cerebro = bt.Cerebro()
cerebro.addstrategy(SmaCross)

data0 = bt.feeds.YahooFinanceData(dataname='MSFT', fromdate=datetime(2011, 1, 1),
                                  todate=datetime(2012, 12, 31))
cerebro.adddata(data0)

cerebro.run()
cerebro.plot()
```

```python
import backtrader as bt
from matplotlib import pyplot as plt

class SmaCross(bt.Strategy):
    params = dict(pfast=5, pslow=20)
    def __init__(self):
        sma_fast = bt.ind.SMA(period=self.p.pfast)
        sma_slow = bt.ind.SMA(period=self.p.pslow)
        self.crossover = bt.ind.CrossOver(sma_fast, sma_slow)

    def next(self):
        if not self.position:
            if self.crossover > 0:
                self.buy()
        elif self.crossover < 0:
            self.close()

cerebro = bt.Cerebro()
cerebro.adddata(bt.feeds.GenericCSVData(dataname='datas/test01.csv', dtformat='%Y-%m-%d'))
cerebro.addstrategy(SmaCross)
cerebro.broker.setcash(10000.0)
cerebro.broker.setcommission(commission=0.001)
cerebro.run()
cerebro.plot()
# plt.show()
```

```python
import edge_tts
import asyncio

async def save_text_to_mp3(text: str, output_file: str = "output.mp3"):
    """
    使用 edge-tts 将文本合成为语音并保存为 MP3 文件。
    """
    # 1. 创建 Communicate 对象
    # 你可以尝试不同的语音，例如一个更正式的男声 "zh-CN-YunxiNeural"
    communicate = edge_tts.Communicate(text, voice="zh-CN-YunxiNeural")

    print(f"正在将文本合成语音并保存到: {output_file}")

    # 2. 使用 save 方法保存到文件
    await communicate.save(output_file)

    print(f"语音已成功保存到文件: {output_file}")

# --- 主程序 ---
if __name__ == "__main__":
    text_to_speak = "2026年我家牛牛就要一岁啦, 希望能做一个智能机器人, 伴随宝宝一起健康成长, 一起跑跑跳跳长高高."
    output_filename = "my_speech.mp3"
    
    await save_text_to_mp3(text_to_speak, output_filename)
```
