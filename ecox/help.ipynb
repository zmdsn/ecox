```python
import numpy as np

def estimate_price_by_roe(bps0, roe, ke, years=5, g=0.02):
    """
    使用简化剩余收益模型估算股价
    参数：
    bps0: 当前每股净资产
    roe: 预测的未来ROE（假设前几年恒定）
    ke: 股东必要报酬率
    years: 高速预测年数
    g: 永续增长率
    """
    estimated_ri_pv = 0  # 剩余收益现值
    bps = bps0

    # 计算预测期内的剩余收益现值
    for t in range(1, years + 1):
        ri = bps * (roe - ke)  # 当年剩余收益
        ri_pv = ri / ((1 + ke) ** t)  # 折现
        estimated_ri_pv += ri_pv
        bps = bps * (1 + roe)  # 假设利润全留存，净资产增长

    # 计算永续终端价值（预测期后）
    ri_terminal = bps * (roe - ke) * (1 + g)
    terminal_value_pv = ri_terminal / ((ke - g) * ((1 + ke) ** years))

    # 内在价值 = 当前净资产 + 剩余收益现值总和
    intrinsic_value = bps0 + estimated_ri_pv + terminal_value_pv
    return intrinsic_value

# 示例：假设一家公司
bps0 = 6.91 # 当前每股净资产10元
roe = 0.17  # 预测未来ROE为15%
ke = 0.10    # 股东要求10%的回报率
g = 0.04
estimated_price = estimate_price_by_roe(bps0, roe, ke, g=g)
print(f"基于模型估算的每股内在价值约为：{estimated_price:.2f} 元")
```

```python
# df['ROE'].plot()
# df['REPORT_DATE_NAME']

df['AVG_TOTAL_ASSETS'] = (df['TOTAL_ASSETS'] + df.groupby('SECURITY_CODE')['TOTAL_ASSETS'].shift(1)) / 2
df['AVG_TOTAL_EQUITY'] = (df['TOTAL_EQUITY'] + df.groupby('SECURITY_CODE')['TOTAL_EQUITY'].shift(1)) / 2

# 2. 计算杜邦分析三大比率
# 净利率 (Profit Margin) = 净利润 / 营业收入
# 使用“归属于母公司净利润”和“营业收入”
df['Net_Profit_Margin'] = df['PARENT_NETPROFIT'] / df['OPERATE_INCOME']

# 总资产周转率 (Asset Turnover) = 营业收入 / 平均总资产
df['Asset_Turnover'] = df['OPERATE_INCOME'] / df['AVG_TOTAL_ASSETS']

# 权益乘数 (Equity Multiplier) = 平均总资产 / 平均净资产
df['Equity_Multiplier'] = df['AVG_TOTAL_ASSETS'] / df['AVG_TOTAL_EQUITY']

# 3. 计算ROE（作为验证和对比）
df['ROE_by_Dupont'] = df['Net_Profit_Margin'] * df['Asset_Turnover'] * df['Equity_Multiplier']
```
